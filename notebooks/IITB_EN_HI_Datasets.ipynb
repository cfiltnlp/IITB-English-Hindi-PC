# **Format Data**


```python
import json

source_train = open("train.en").readlines()
target_train = open("train.hi").readlines()

source_valid = open("dev_test/dev.en").readlines()
target_valid = open("dev_test/dev.hi").readlines()

source_test = open("dev_test/test.en").readlines()
target_test = open("dev_test/test.hi").readlines()

data_train = []
data_test = []
data_valid = []

for (src, tgt) in zip(source_train, target_train):
  output = {}
  translation = {}
  translation["en"] = src.strip("\n")
  translation["hi"] = tgt.strip("\n")
  output["translation"] = translation
  data_train.append(output)

for (src, tgt) in zip(source_valid, target_valid):
  output = {}
  translation = {}
  translation["en"] = src.strip("\n")
  translation["hi"] = tgt.strip("\n")
  output["translation"] = translation
  data_valid.append(output)

for (src, tgt) in zip(source_test, target_test):
  output = {}
  translation = {}
  translation["en"] = src.strip("\n")
  translation["hi"] = tgt.strip("\n")
  output["translation"] = translation
  data_test.append(output)

json_train = json.dumps({"data": data_train}, ensure_ascii=False)
json_test = json.dumps({"data": data_test}, ensure_ascii=False)
json_valid = json.dumps({"data": data_valid}, ensure_ascii=False)

with open("train.json", "w+", encoding='utf8') as f:
  f.write(json_train)
with open("valid.json", "w+", encoding='utf8') as f:
  f.write(json_valid)
with open("test.json", "w+", encoding='utf8') as f:
  f.write(json_test)
```
